```python
# program of matplotlib in python
import matplotlib.pyplot as p
import numpy as np
x = np.array([3,8])
y = np.array([9,6])
p.plot(x,y)
p.show()
```

```python
import matplotlib.pyplot as p
data = (2,5,1,4)
p.plot(data)
p.show()
```

```python
import matplotlib.pyplot as p
import numpy as np
a = np.arange (0,2*np.pi,0.1)
b = np.sin(a)
p.plot(a,b)
p.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
 
# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)
 
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
plt.boxplot(data)
 
# show plot
plt.show()
```

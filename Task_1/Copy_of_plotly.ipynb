```python
import plotly.express as px
df = px.data.tips()
fig = px.pie(df, values='tip',names='day')
fig.show()
```

```python
import plotly.express as px
data_india = px.data.gapminder()
("country == 'India'")
fig = px.bar(data_india, x= 'year',y= 'pop')
fig.show() 
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x="total_bill")
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007")

fig = px.scatter(df, x="gdpPercap", y="lifeExp", hover_name="country",
                 log_x=True, range_x=[1,10000], range_y=[0,100])
fig.show()
```
